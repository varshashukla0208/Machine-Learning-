##### to store passenger data of the Titanic. For a number of passengers, I know the name (characters), age (integers) and sex (male/female) data.

```python
df = pd.DataFrame({
                    "name": ['Pooja', "Ragani", "Anjali", "Sonu", "Shubham"],
                    "age":  [14, 24, 29, 31, 32],
                    "sex ":  ['Female', 'Female','Female', 'Male', 'Male']    
})

df
```

```python
#### Finding the Maximum age of passengers.
max_age = df['age'].max()
print(max_age)
```

##### The describe() method provides a quick overview of the numerical data in a DataFrame. As the Name and Sex columns are textual data, these are by default not taken into account by the describe() method

```python
df.describe()
```

## How do I select a subset of a DataFrame?

#### I’m interested in the age of the Titanic passengers.

```python
ages = titanic['Age']
```

```python
ages.head()
type(ages)
titanic.shape
titanic['Age'].shape
```

#### To select multiple columns, use a list of column names within the selection brackets [].

```python
multi = titanic[['Age', 'Sex']]
multi.head(10)
multi.shape
```

### How do I filter specific rows from a DataFrame?

```python
### I’m interested in the passengers older than 35 years.

above_age = titanic[titanic["Age"] > 35]
above_age.head()
```

```python
titanic["Age"] > 35
```

```python
above_age.shape
```

```python
### I’m interested in the Titanic passengers from cabin class 2 and 3.

class_23 = titanic[titanic["Pclass"].isin([2, 3])]
```
