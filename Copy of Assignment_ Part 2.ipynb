### **Q6. Find Maximum in a List**

**Description:**

Take `N` integers in a list and print the maximum number.

We can store a given `N` Integer in a list name `numbers`.

Note:

- We can take input from the user using `input()`.
- We can also directly take input as an Integer from the user using `int(input())`.

**For example, Input:**

```

  
numbers = [10, 3, 25, 8, 7]

```

**Output:**

```
  
25

```

---



````python
n = int(input("Enter the range of number you want: "))

numbers = []


for i in range(n):
  num = int(input(f"Enter number {i+1}: "))
  numbers.append(num)
print("The list of numbers is:", numbers)
print("The maximum number among the given list is: ",max(numbers))
````

### **Q7. Reverse a List**

**Description:**

Take a list of numbers and print it in reverse order.

We can store a given `N` numbers in a list name `numbers`.

Note:

- We can take input from the user using `input()`.
- We can also directly take input as an Integer from the user using `int(input())`.

**For example, Input:**

```

  
numbers = [1, 2, 3, 4]

```

**Output:**

```
csharp
  
[4, 3, 2, 1]

```

---



````python
n = int(input("Enter the range for numbers: "))

numbers = []

for i in range(n):
  num = int(input(f"Enter the number for list{i+1}: "))
  numbers.append(num)
print("Original List: ", numbers)
print("Reversed list: ",numbers[: : -1])

````
