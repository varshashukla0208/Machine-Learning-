```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cardata.csv')
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

### Scatter Plot

```python
plt.figure(figsize=(8,6))
#plt.legend()
print(df.head())

# Step 2: Plot the data (example: line plot for two columns)
plt.scatter(df['Present_Price'], df['Selling_Price'], color='blue', alpha=0.7)
plt.title('Selling Price vs Present Price')
plt.xlabel('Present Price (in lakhs)')
plt.ylabel('Selling Price (in lakhs)')
plt.grid(True)
plt.tight_layout()
plt.show()


```

#### Bar Plot

```python
df['Selling_Price'].value_counts().plot(kind='bar')
plt.title('Bar Chart of Column2')
plt.show()
```

#### Histogram

```python
df['Present_Price'].plot(kind='hist', bins=10)
plt.title('Histogram of Column1')
plt.show()
```

#### Line plot

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Present_Price'], df['Selling_Price'], marker='o', linestyle='-', color='green')
plt.title('Line Plot of Selling Price vs Present Price')
plt.xlabel('Present Price (in lakhs)')
plt.ylabel('Selling Price (in lakhs)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Bar Plot

```python
plt.bar(df['Present_Price'], df['Selling_Price'], color='skyblue', edgecolor='black')

# Labels and title
plt.legend() 
plt.title('Selling Price for Each Present Price')
plt.xlabel('Present Price (in lakhs)')
plt.ylabel('Selling Price (in lakhs)')
plt.xticks(rotation=90)  # Rotate x-axis labels to prevent overlap
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

#### Scatter Plot

```python
plt.scatter(df['Present_Price'], df['Selling_Price'], 
            color='orange', alpha=0.7, label='Selling Price vs Present Price')

# Add labels, title, and legend
plt.title('Scatter Plot: Selling Price vs Present Price')
plt.xlabel('Present Price (in lakhs)')
plt.ylabel('Selling Price (in lakhs)')
plt.legend()  # <--- Add legend here
plt.grid(True)
plt.tight_layout()
plt.show()
```
