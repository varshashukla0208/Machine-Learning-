#### Find the average Age of survivors vs non-survivors.

```python
avr_survived = df.groupby('Survived')['Age'].mean()
avr_survived
```

#### Find survival rate by Sex.

```python
avg_rate = df.groupby('Sex')['Survived'].mean()
print(avg_rate)
```

### Find survival rate by Pclass.

```python
pass1 = df.groupby('Pclass')['Survived'].mean()
pass1
```

### Find survival rate by both Sex and Pclass.

```python
avg_survive = df.groupby(['Sex', 'Pclass'])['Survived'].mean()
avg_survive
```

### Fill missing Age values with the median Age.

```python
age_fill = df['Age'].fillna(df['Age'].median(), inplace=True)
```

### Fill missing Embarked values with the most common port.

```python
embark_col = df['Embarked'].mode()[0]
df.fillna(embark_col, inplace=True)
df['Embarked'].isna().sum()
```

### Create a new column “FamilySize” = SibSp + Parch + 1.

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df['FamilySize'].head()
```

#### Create a new column “IsAlone” (1 if FamilySize=1 else 0).

```python
df['IsAlone'] = (df['FamilySize'] == 1 ).astype(int)
df['IsAlone']
```
